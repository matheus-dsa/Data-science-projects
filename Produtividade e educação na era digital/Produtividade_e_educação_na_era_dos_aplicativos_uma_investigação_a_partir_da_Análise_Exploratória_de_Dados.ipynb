# **Sobre os dados**

O dataset foi obtido na plataforma Kaggle:

https://www.kaggle.com/lava18/google-play-store-apps

São dados sobre cada aplicativo para android na plataforma do Google Play Store

**Dicionário de variáveis**
=======================

---
**App**: Nome do aplicativo

**Category**: Categoria dos aplicativos

**Rating**: Nota média de cada aplicativo

**Reviews:** Quantidade de notas que o aplicativo recebeu

**Size:** Tamanho do aplicativo

**Installs:** Quantas pessoas instalaram o aplicativo

**Type:** Se é gratuito ou pago

**Price:** Quanto custa

**Content Rating:** Classificação etária mínima

**Genres:** Gênero do aplicativo (similar à Categoria)

**Last Updated:** Última atualização

**Current ver:** Versão atual

**Android ver:** Versão do Android necessária para download

# **Mãos no código!!**

````
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
import pylab
pylab.rcParams['figure.figsize'] = (10, 5)
````

````
df = pd.read_csv('googleplaystore.csv')
df.head()
````

````
df.info()
````

**Verificando os dados**
========================

Sem dúvidas, a parte mais importante da análise é a limpeza e organização dos dados, sem ela, nossos dados continuam embaralhados e confusos

data.png

````
df.Category.unique()
````

````
df['Category'].value_counts(1)
````

Limitando nosso datafreme para Educação e Produtividade.

Farei isso porque é importante direcionar nossa análise. Aqui eu busco entender como eles são avaliados no mercado de apps, então preço, versão, tamanho e mais alguns não são tão interessantes assim, portanto é preferível excluí-los

Primeiramente, excluindo as colunas indesejadas

````
df_educ = df[df.Category == 'EDUCATION']
df_educ = df_educ.drop(['Genres','Size', 'Price', 'Last Updated', 'Current Ver', 'Android Ver', 'Content Rating'], axis=1)

df_educ.head()
````

Podemos até observar alguns aplicativos educacionais bem conhecidos, como **Duolingo** (estudar idiomas), **TED** (palestras sobre diversos assuntos) e **Khan Academy** (plataforma com aulas de diversas áreas diferentes)

Excluindo as colunas indesejadas para o dataframe dos apps de produtividade

````
df_product = df[df.Category=='PRODUCTIVITY']
df_product = df_product.drop(['Genres','Size', 'Price', 'Last Updated', 'Current Ver', 'Android Ver', 'Content Rating'], axis=1)

df_product.head()
````

Aqui também podemos observar alguns aplicativos de produtividade conhecidos, como **Microsoft Word** (editor de textos), **Adobe Acrobat Reader** (Leitor de PDF), **Google Drive** (Armazenamento na nuvem)

````
print('Educação')
pd.to_numeric(df_educ.Reviews)
df_educ.Reviews = df_educ.Reviews.astype(int)
print(f'''Total de avaliações: {df_educ.Reviews.sum()}
Média das avaliações: {df_educ.Reviews.mean():.2f}''')
````

````
df_educ = df_educ.sort_values('Reviews', ascending=False)
df_educ = df_educ.drop_duplicates('App')
df_educ = df_educ.dropna()
````

````
df_product.sort_values('Reviews', ascending=False)
df_product.dropna(inplace=True)
df_product.loc[df_product['Reviews']=='5']
````

Agora que já limpamos os dados, iremos concatená-los, formando um só, assim criamos o dataframe que iremos analisar a partir de agora

````
df_educ_product = pd.concat([df_educ, df_product])
df_educ_product
````

Fazer as descrições estatísticas do nosso dataframe para entender como estão distribuídas suas principais características estatísticas, como: quantidade, média, desvio padrão, valor mínimo, valor máximo e quartis (primeiro, segundo e terceiro)

````
df_educ_product.describe()
````

````
df_educ_product.Reviews = df_educ_product.Reviews.astype(int)
````

````
notas = pd.DataFrame(df_educ_product.groupby('App')['Rating'].mean())
notas.dropna(inplace=True)
notas['número de avaliações'] = pd.DataFrame(df_educ_product.groupby('App')['Reviews'].count())
notas = df_educ_product[df_educ_product['Reviews'] > 10]
notas
````

Limpar os dados de instalações, pois como podemos observar, temos uma vírgula separando os milhares, e um símbolo de + no final

````
df_educ_product.Installs = df_educ_product.Installs.apply(lambda x: x.strip('+'))
df_educ_product.Installs = df_educ_product.Installs.apply(lambda x: x.replace(',',''))
df_educ_product.Installs = df_educ_product.Installs.astype(int) 
df_educ_product.info()
````

Faremos o mesmo processo nos dataframes de educação e de produtividade, pois os utilizaremos no final

````
df_educ['Installs'] = df_educ['Installs'].apply(lambda x: x.strip('+'))
df_educ.Installs = df_educ.Installs.apply(lambda x: x.replace(',',''))
df_educ.Installs = df_educ.Installs.astype(int)

df_product.Installs = df_product.Installs.apply(lambda x: x.strip('+'))
df_product.Installs = df_product.Installs.apply(lambda x: x.replace(',',''))
df_product.Installs = df_product.Installs.astype(int)
````

AGora colocaremos os dados em ordem, através da coluna "Reviews", que se refere à quantidade de avaliações. Os mais avaliados ficam no início, e os menos avaliados no final. Depois disso, retiramos os dados duplicados

````
notas.sort_values('Reviews', ascending=False, inplace= True)
notas = notas['App'].drop_duplicates()
notas
````

Verificando as informações sobre nosso dataframe

````
df_educ_product.info()
````

# **Produtividade e educação**

---

Agora iremos análisar e comparar os dados estatísticos sobre os aplicativps de produtividade e os aplicativos de educação, para que possamos ter uma noção de como estão distribuídos na loja virtual.

Aqui começa nossa **Análise Exploratória de Dados**

graph.png

````

Produtividade = df_educ_product['Category'].value_counts()[0]
Educação = df_educ_product['Category'].value_counts()[1]

print(f'Total: {Educação + Produtividade} ')
print('Produtividade(%): {:.2f}'.format(Produtividade/ (Educação + Produtividade)*100))
print('Educação(%): {:.2f}'.format(Educação / (Educação + Produtividade)*100))
````

````
sns.countplot('Category', data=df_educ_product)
plt.title('Aplicativos de educação e produtividade', fontsize=17)
plt.ylabel('Quantidade', fontsize=13)
plt.xlabel('Categoria',fontsize=13);
````

O número de apps de produtividade é muito maior que os de educação, aproximadamente **75%** contra apenas **25%**.

````
app_pago = df_educ_product[df_educ_product['Type'] == 'Paid']['Category'].value_counts()
produtividade_pago = app_pago[0]
educação_pago = app_pago[1]
app_gratis = df_educ_product[df_educ_product['Type'] == 'Free']['Category'].value_counts()
produtividade_gratis = app_gratis[0]
educação_gratis = app_gratis[1]
app_gratis_total = produtividade_gratis+educação_gratis
app_pago_total = produtividade_pago + educação_pago
````
